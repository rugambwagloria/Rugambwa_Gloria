```python
import pandas as pd
df=pd.read_csv('AfricaCupOfNationsMatches.csv')
```

```python
df.head(7)
```

```python
cols=['HomeTeam','AwayTeam','HomeTeamGoals','AwayTeamGoals']
df=df[cols]
print(df.to_string)
```

```python

# Filter rows where 'HomeTeam' or 'AwayTeam' is 'Egypt'
egypt_matches = df[(df['HomeTeam'] == 'Egypt') | (df['AwayTeam'] == 'Egypt')]
egypt_matches
```
